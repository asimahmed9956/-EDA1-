```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogram 
plt.hist(df['Unit price'], bins=20)
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.title('Distribution of Product Prices')
plt.show()

# Bar plot 
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Product line', hue='Gender')
plt.xticks(rotation=45)
plt.title('Distribution of Product Lines by Gender')
plt.show()

# Scatter plot 
plt.scatter(df['Quantity'], df['Total'])
plt.xlabel('Quantity')
plt.ylabel('Total')
plt.title('Quantity vs Total')
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Box plot of Gross Income by Payment type
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Payment', y='gross income')
plt.title('Gross Income by Payment Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Line plot of total sales over time
plt.figure(figsize=(10, 6))
df.groupby('Date')['Total'].sum().plot()
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Total Sales Over Time')
plt.show()
```
